```python
df = df.drop(columns=['date', 'lights'])
```

```python
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
normalised_df
```

```python
x = normalised_df.iloc[:,1:]
y = normalised_df['Appliances']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
# obtain predictions
predicted_values = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 3) 
```

```python
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3) 
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
def get_weight(model, feature, col_name):
    weights = pd.Series(model.coef_, feature.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_weights = get_weight(model, x_train, 'Weight')
```

```python
linear_weights
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
predicted_val = ridge_reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_val))
round(rmse, 3) 
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
predicted_va = lasso_reg.predict(x_test)
```

```python
lasso_weights = get_weight(lasso_reg, x_train, 'Lasso_weight')
```
